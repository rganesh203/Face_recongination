```python
import numpy as np
import cv2
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("E:\IMG-7458.JPG")
img.shape
#print(img)
```

```python
plt.imshow(img)
```

```python
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
gray
```

```python
plt.imshow(gray, cmap='gray')
```

```python
face= cv2.CascadeClassifier('./Downloads/haarcascade_frontalface_default.xml')
face= face.detectMultiScale(gray, 1.3, 5)
face
```

```python
(x,y,w,h)= face[0]
x,y,w,h
```

```python
face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
plt.imshow(face_img)
```
